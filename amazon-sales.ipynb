```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To check the proportion of missing values
missing_values = df.isnull().sum() / len(df) * 100
missing_values
```

```python
#To remove comma
df['rating_count'] = df['rating_count'].str.replace(',', '')
```

```python
df.rating_count
```

```python
#To fill NaN with 0
df['rating_count'] = df['rating_count'].str.replace(',', '').fillna(0)
#To Drop rows with NaN values in 'rating_count'
#df.dropna(subset=['rating_count'], inplace=True)
```

```python
#To convert string to integer
df['rating_count'] = df['rating_count'].astype(int)
```

```python
#Filling missing value with mean
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].mean())
```

```python
df.dtypes
```

```python
# To identify non-numeric values
non_numeric_rows = df['rating'].str.isnumeric() == False
print("Non-numeric values:", df[non_numeric_rows])
```

```python
#To replace non-numeric (|) values with NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# Convert column to float
df['rating'] = df['rating'].astype(float)
```

```python
df.dtypes
```

```python
#To remove non-numeric characters(₹)
df['discounted_price'] = df['discounted_price'].replace('[^0-9]', '', regex=True)
```

```python
df.discounted_price
```

```python
#To convert cleaned column to integer
df['discounted_price'] = df['discounted_price'].astype(int)
```

```python
df.dtypes
```

```python
#To remove non-numeric characters (₹ and commas)
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True)
```

```python
#To convert string to float
df['actual_price'] = df['actual_price'].astype(float)
```

```python
#To remove the '%' symbol
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(int)
```

```python
df.duplicated().sum()
```

# **Data Analysis & Visualization**

```python
#Total revenue for each product
df['total_revenue'] = df['discounted_price'] * df['rating_count']
df.total_revenue
```

```python
# Top 10 products by total revenue
top_10_products = df.groupby('product_name')['total_revenue'].sum().nlargest(10).reset_index()
top_10_products
```

```python
# Sales distribution by category
category_sales = df.groupby('category')['total_revenue'].sum().nlargest(10).reset_index()
category_sales.sort_values(by ="total_revenue", ascending = False)
```

```python
# Heatmap of correlations
corr_matrix = df[['discount_percentage', 'rating_count', 'total_revenue']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Number of ratings per category
category_ratings = df.groupby('category')['rating_count'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='rating_count', y='category', data=category_ratings, palette='viridis')
plt.title('Number of Ratings per Category')
plt.xlabel('Rating Count')
plt.ylabel('Category')
plt.show()
```

```python
# Suppress the specific FutureWarning
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# Replace inf values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

#Distribution of ratings across products
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=10, kde=True, color='green')
plt.title('Distribution of Ratings Across Products')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Word Cloud for most frequently mentioned keywords in reviews
from wordcloud import WordCloud

review_text = " ".join(review for review in df['review_content'])
wordcloud = WordCloud(background_color='white', max_words=100, contour_width=3, contour_color='steelblue').generate(review_text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from textblob import TextBlob

# Sentiment analysis on reviews
df['sentiment_score'] = df['review_content'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Average sentiment score per category
category_sentiment = df.groupby('category')['sentiment_score'].mean().nlargest(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='sentiment_score', y='category', data=category_sentiment, palette='viridis')
plt.title('Average Sentiment Score per Category')
plt.xlabel('Sentiment Score')
plt.ylabel('Category')
plt.show()
```

```python
# Sentiment distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['sentiment_score'], bins=20, kde=True, color='purple')
plt.title('Sentiment Score Distribution')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Number of reviews per category
category_reviews = df.groupby('category')['review_id'].count().nlargest(5).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='review_id', y='category', data=category_reviews, palette='viridis')
plt.title('Number of Reviews per Category')
plt.xlabel('Number of Reviews')
plt.ylabel('Category')
plt.show()
```

```python
# Word Cloud for common words in reviews
review_text = " ".join(review for review in df['review_content'])
wordcloud = WordCloud(background_color='white', max_words=100, contour_width=3, contour_color='steelblue').generate(review_text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Number of reviews per user
user_reviews = df.groupby('user_id')['review_id'].count().nlargest(7).reset_index()
user_reviews.sort_values(by = "review_id", ascending = False)
```

```python
# Percentage of products with valid image links
df['img_valid'] = df['img_link'].apply(lambda x: pd.notnull(x) and x.startswith('http'))
valid_img_percentage = df['img_valid'].mean() * 100
plt.figure(figsize=(6, 6))
plt.pie([valid_img_percentage, 100 - valid_img_percentage], labels=['Valid', 'Invalid'], autopct='%1.1f%%', colors=sns.color_palette('viridis'))
plt.title('Percentage of Products with Valid Image Links')
plt.show()
```

```python
# Average rating by category
category_rating = df.groupby('category')['rating'].mean().nlargest(6).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='category', data=category_rating, palette='viridis')
plt.title('Average Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```
